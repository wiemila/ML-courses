```python
dataXbezVal = df2[['DB 5', 'DB 4 ', 'Kat 3', 'Kat 5', 'NH4', 'A5Cl','A10Cl', 'NH42']].values
dataXbezVal
pca = PCA(n_components=2)
pca.fit(dataXbezVal)
print(pca.explained_variance_ratio_)
x_pca = pca.transform(dataXbezVal)
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=dataY, s = 100,
            alpha = 0.8, marker = '.')
plt.title('PC 1 vs PC 2')
plt.xlabel('PC 1 (%.2f%%)' % (pca.explained_variance_ratio_[0]*100))
plt.ylabel('PC 2 (%.2f%%)' % (pca.explained_variance_ratio_[1]*100))

plt.colorbar();
```

```python
name = df2[['DB 5', 'DB 4 ', 'Kat 3', 'Kat 5', 'NH4', 'A5Cl','A10Cl', 'NH42']]
plt.scatter(pca.components_[0,:],pca.components_[1,:],cmap='Set1')
plt.title('Loadings PC 1 vs PC 2')
plt.xlabel('LoadingsPC 1 (%.2f%%)' % (pca.explained_variance_ratio_[0]*100))
plt.ylabel('LoadingsPC 2 (%.2f%%)' % (pca.explained_variance_ratio_[1]*100))
for i, txt in enumerate(name):
    plt.annotate(txt, (pca.components_[0,:][i],pca.components_[1,:][i]))
plt.colorbar();
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=0, n_init="auto").fit(dataXbezVal)
y_kmeans = kmeans.predict(dataXbezVal)
```

```python
plt.scatter(x_pca[:,0],x_pca[:,1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```
